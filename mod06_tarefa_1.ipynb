#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import seaborn as sns
```

```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://diaad.s3.sa-east-1.amazonaws.com/sinasc/SINASC_2019.csv"
df = pd.read_csv(url, sep = ';', decimal = ',')
```

```python
roraima_df = df[df['CODUFNATU'] == 11.0]
roraima_df.ESTCIVMAE
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python

variaveis = ['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE']

for var in variaveis:
    plt.figure(figsize=(10, 5))

    # Usando a função countplot do Seaborn para criar um gráfico de barras.
    # data=roraima_df: especifica o DataFrame de onde os dados serão retirados.
    # x=var: indica que a coluna atual do loop será usada no eixo x.
    # palette="viridis": escolhe uma paleta de cores específica para as barras.
    sns.countplot(data=roraima_df, x=var, palette="viridis")

    # Definindo o título do gráfico.
    plt.title(f'Distribuição de {var}')

    # Definindo o rótulo do eixo y.
    plt.ylabel('Quantidade')

    plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python

plt.figure(figsize=(10, 5))

# Gráfico de barras: média de 'APGAR5' agrupada por 'CONSULTAS', com desvio padrão como barra de erro
sns.barplot(x='CONSULTAS', y='APGAR5', data=roraima_df, ci='sd', palette='viridis')

# Título do gráfico
plt.title('APGAR5 médio por CONSULTAS')

# Rótulos dos eixos
plt.ylabel('APGAR5 Médio')
plt.xlabel('CONSULTAS')

plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python

fig, ax = plt.subplots(2, 2, figsize=(15, 10))

# Definir um título geral para toda a figura.
fig.suptitle('Painel de frequencias')

# Contar a frequência de cada valor na coluna 'SEXO' e criar um gráfico de barras.
# ax=ax[0, 0] determina que este gráfico será posicionado na primeira linha e primeira coluna do painel.
roraima_df['SEXO'].value_counts().plot(kind='bar', ax=ax[0, 0], color='skyblue')

# Definir o título do gráfico na posição ax[0, 0].
ax[0, 0].set_title('SEXO')

# Definir o rótulo do eixo y para o gráfico na posição ax[0, 0].
ax[0, 0].set_ylabel('Frequência')

# Similar ao gráfico 'SEXO', mas para a coluna 'GESTACAO' e posicionado na primeira linha e segunda coluna.
roraima_df['GESTACAO'].value_counts().plot(kind='bar', ax=ax[0, 1], color='salmon')
ax[0, 1].set_title('GESTACAO')

# Similar aos gráficos anteriores, mas para a coluna 'ESTCIVMAE' e posicionado na segunda linha e primeira coluna.
roraima_df['ESTCIVMAE'].value_counts().plot(kind='bar', ax=ax[1, 0], color='lightgreen')
ax[1, 0].set_title('ESTCIVMAE')
ax[1, 0].set_ylabel('Frequência')

# Similar aos gráficos anteriores, mas para a coluna 'CONSULTAS' e posicionado na segunda linha e segunda coluna.
roraima_df['CONSULTAS'].value_counts().plot(kind='bar', ax=ax[1, 1], color='lightcoral')
ax[1, 1].set_title('CONSULTAS')

# Ajusta o layout dos subgráficos para que não haja sobreposição.
plt.tight_layout()

# Ajusta a posição dos subgráficos para que não interfiram no título geral.
plt.subplots_adjust(top=0.90)


plt.show()
```
